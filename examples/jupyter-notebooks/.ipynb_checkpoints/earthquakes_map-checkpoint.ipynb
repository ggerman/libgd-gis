```ruby
# ═════════════════════════════════════════════════════════════════
#  Earthquakes Map - Depth Visualization
#  Author: Germán Alberto Giménez Silva <ggerman@gmail.com>
#  Library: libgd-gis / ruby-libgd
#  Description: Recent earthquakes (M4.5+) from USGS.
#               Red = shallow (<70km), Blue = deep (≥70km).
#               Circle size indicates magnitude.
# ═════════════════════════════════════════════════════════════════

require 'gd'
require 'gd/gis'
require 'net/http'
require 'json'

def render(file_name)
    IRuby.display "<img src='#{file_name}' />", mime: "text/html"
end

# Polaroid frame helper
def polaroid_frame(image, margin: 20, bottom_margin: 80, bg_color: [255, 255, 255, 255])
  new_width  = image.width + margin * 2
  new_height = image.height + margin + bottom_margin

  framed = GD::Image.new(new_width, new_height)
  framed.antialias = false

  bg = GD::Color.rgba(*bg_color)
  framed.filled_rectangle(0, 0, new_width, new_height, bg)
  framed.copy(image, margin, margin, 0, 0, image.width, image.height)
  framed
end

# Fetch earthquake data
puts "🌍 Fetching earthquake data (M4.5+)..."

url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/4.5_month.geojson"
response = Net::HTTP.get(URI(url))
data = JSON.parse(response)
earthquakes = data['features']

valid_quakes = earthquakes.select { |eq| eq['geometry']['coordinates'][1].between?(-85, 85) }
puts "✅ Retrieved #{valid_quakes.size} significant earthquakes"

# Create map
map = GD::GIS::Map.new(
  bbox: [-180, -85, 180, 85],
  zoom: 2,
  basemap: :carto_light,
  width: 1000,
  height: 600
)

map.style = GD::GIS::Style.load("light")
font_path = GD::GIS::FontHelper.find("DejaVuSans") || GD::GIS::FontHelper.random

# Ensure points have font
map.style.points[:font] ||= font_path
map.style.points[:size] ||= 12

# Shallow earthquakes (depth < 70km) - Red
shallow = valid_quakes.select { |eq| eq['geometry']['coordinates'][2] < 70 }
shallow_data = shallow.map do |eq|
  coords = eq['geometry']['coordinates']
  { lon: coords[0], lat: coords[1], label: "#{eq['properties']['mag'].round(1)}" }
end

if shallow_data.any?
  layer = GD::GIS::PointsLayer.new(
    shallow_data,
    lon: ->(p) { p[:lon] }, lat: ->(p) { p[:lat] },
    icon: "numeric", label: ->(p) { p[:label] },
    font: font_path, size: 13,
    color: [220, 70, 70, 220],
    font_color: [255, 255, 255, 255]
  )
  map.instance_variable_get(:@points_layers) << layer
  puts "📍 Shallow earthquakes (<70km): #{shallow.size}"
end

# Deep earthquakes (depth >= 70km) - Blue
deep = valid_quakes.select { |eq| eq['geometry']['coordinates'][2] >= 70 }
deep_data = deep.map do |eq|
  coords = eq['geometry']['coordinates']
  { lon: coords[0], lat: coords[1], label: "#{eq['properties']['mag'].round(1)}" }
end

if deep_data.any?
  layer = GD::GIS::PointsLayer.new(
    deep_data,
    lon: ->(p) { p[:lon] }, lat: ->(p) { p[:lat] },
    icon: "numeric", label: ->(p) { p[:label] },
    font: font_path, size: 13,
    color: [70, 100, 220, 220],
    font_color: [255, 255, 255, 255]
  )
  map.instance_variable_get(:@points_layers) << layer
  puts "📍 Deep earthquakes (≥70km): #{deep.size}"
end

# Render
map.render

# Add polaroid frame
framed = polaroid_frame(map.image, margin: 20, bottom_margin: 90)

# Add title and description
framed.text_ft(
  "🌍 Global Earthquakes (M4.5+) — Last 30 Days",
  x: 30, y: framed.height - 70,
  font: font_path, size: 14,
  color: GD::Color.rgb(40, 40, 40)
)

framed.text_ft(
  "🔴 Shallow (<70km)  |  🔵 Deep (≥70km)  |  Circle size = Magnitude | Data: USGS",
  x: 30, y: framed.height - 50,
  font: font_path, size: 11,
  color: GD::Color.rgb(80, 80, 80)
)

# Add author and library credit
framed.text_ft(
  "Germán A. Giménez Silva <ggerman@gmail.com> | libgd-gis / ruby-libgd",
  x: 30, y: framed.height - 30,
  font: font_path, size: 10,
  color: GD::Color.rgb(120, 120, 120)
)

# Save
framed.save("/work/earthquakes_framed.png")
render("earthquakes_framed.png")
puts "✅ Saved: earthquakes_framed.png"
```
